## 1. Working with Interview Data
## 1.1 The `with` Statement

Up until now, we’ve worked only with data created inside our own code. But in real-world projects, you’ll mainly work with existing data and files. 

Many qualitative researchers store interview transcripts or field notes as .docx files (Microsoft Word). However, for computational text analysis, it is usually best to work with plain text files (.txt). A .docx file does not only contain text, it also stores formatting information (fonts, styles, layout, etc.). For most forms of text analysis, we are only interested in the textual content itself, not the formatting. Therefore, it is good practice to convert .docx files into .txt files before starting your analysis.

The following example shows how you can extract the text from a .docx file and save it as a .txt file using the [python-docx library](https://python-docx.readthedocs.io/en/latest/).  We then use the `with` statement to open the data in Python. It is a tool built into Python.

For this workshop, we generated four fictional interviews in order to have realistic data. The people and information are fictitious. The transcripts were generated by GPT-5.2 in 2026/03 for training purposes.

````python
# Import 'python-docx' library
from docx import Document

# Load the .docx file
doc = Document("../Data/DataA/interviews/raw/interview_01.docx")

# Create an empty list to store the paragraph texts
paragraph_texts = []

# Go through each paragraph in the document
for para in doc.paragraphs:
    
    # Extract the text of the paragraph
    paragraph_texts.append(para.text)

# Join all paragraph texts into one single string with line breaks
text = "\n".join(paragraph_texts)

# Save as a plain text file
with open("../Data/DataA/interviews/raw/interview_01.txt", "w", encoding="utf-8") as f:
    f.write(text)

# print the first 1000 characters of the text
print(text[:500])
````

If your **text is already in a .txt file,** we can start directly with it. In Python, we first need to read the file, this means telling Python to open the file and bring its content into memory so we can use it in our code.

Think of it like opening a book: you can’t take notes or highlight until the book is open. Similarly, in Python we open the file, store its contents in a variable, and then we can process, analyze, or manipulate the text.

We can use the `with` statement to open a `.txt` file and store its contents in a variable, for example `data`.

In the example here in the notebook, we will now open the file “interview_01.txt” that we just created.

````python
# To open a file (from a relative path) in read mode ("r"):
with open("../Data/DataA/interviews/raw/interview_01.txt", "r") as f: 
    data = f.read()

print(data)
````

**To save text** into a new file, we use almost the same structure, but change the mode to `"w"` (`write`). Note: we haven't changed the text yet; this is just a demonstration.

````python
with open("../Data/DataA/interviews/raw/interview_01.txt", "w", encoding="utf-8") as f:
    f.write(data)  # write the contents of the variable 'data' to the file

# Print the first 100 characters:
print(data[:100])
````

### **Exercise 1: Convert all .docx files to .txt**

So far, we have only converted a single `.docx` file to a `.txt` file. 
To avoid repeating this step manually for each file, we can use a **for-loop** 
to automatically go through the folder and convert each file one by one. 
For this, we use **conditional statements** (see Notebook 04, section 4.3).

**Instructions:**
- Fill in the blanks `___` for the loop (`for`) and the condition (`if`).  
- The rest of the code shows how to read and save files step by step.


````python
from docx import Document  # Import 'python-docx' library to read .docx files
import os  # Import 'os' library to work with folders and file paths

# Path to the folder containing the .docx files
folder_path = "../Data/DataA/interviews/raw"

# List all files in the folder
files = os.listdir(folder_path)

# Loop through all files in the folder
for file in files:                           # <-- Fill in: loop through each file in the list of files
    
    # Check if the current file is a .docx file
    if file.endswith(".docx"):               # <-- Fill in: check if the current file is a .docx file
        
        # Load the .docx file
        doc = Document(folder_path + "/" + file)
        
        # Create an empty list to store paragraph texts
        paragraph_texts = []
        
        # Loop through all paragraphs in the document
        for para in doc.paragraphs:
            
            # Add the text of each paragraph to the list
            paragraph_texts.append(para.text)
        
        # Combine all paragraphs into one single text string
        text = "\n".join(paragraph_texts)
        
        # Create the name for the new .txt file by replacing the .docx extension
        txt_file = folder_path + "/" + file.replace(".docx", ".txt")
        
        # Open the new file and write the extracted text into it
        with open(txt_file, "w", encoding="utf-8") as f:
            f.write(text)
        
        # Print a confirmation message
        print(txt_file + " saved")
````

## 1.2 Load all .txt files into a `pandas` DataFrame

In this step, we will combine all the interview `.txt` files we created earlier into a single table using **pandas**, Python’s library for working with structured data. Each row in the table will represent one interview, and we will store the file name and the text of the interview in separate columns. In combination with **pandas**, we use **os**, a built-in Python library (“standard library”). It allows you to interact with the operating system, for example to access files, folders, and paths in a way that works across different platforms.  

**What happens in this code:**  
1. Python loops through the folder containing the `.txt` files.  
2. Each file is opened and its content is read into memory using the `with` statement.  
3. A dictionary is created for each interview, containing the filename and the text.  
4. All dictionaries are combined into a **DataFrame**, which is similar to a spreadsheet: rows = interviews, columns = filename and text.  

<img src="../Images/dataframe.png" style="width: 300px;">

Our goal here is to get the data ready for preprocessing in the next notebook. We organize the text files into a structured format that is easy to work with for further analysis.

````python
# Import 'pandas' and 'os' libraries
import pandas as pd   # For creating and working with DataFrames
import os             # For accessing files and folders

# Path to the folder containing the .txt files
folder_path = "../Data/DataA/interviews/raw"

# Create an empty list to store each interview's data
interview_data = []

# Loop through all files in the folder
for file in os.listdir(folder_path):
    
    # Only process files that end with ".txt"
    if file.endswith(".txt"):
        
        # Open the text file and read its content
        with open(os.path.join(folder_path, file), "r", encoding="utf-8") as f:
            text = f.read()
        
        # Create a dictionary with filename and text
        data_entry = {
            "filename": file,  # Store the name of the file
            "text": text       # Store the interview text
        }
        
        # Add the dictionary to the list
        interview_data.append(data_entry)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(interview_data)

# Display the first rows of the DataFrame to check the result
df.head()
````

## 2. Collecting Text Data from Wikipedia (Intro to Web Scraping)

In addition to working with interview transcripts, it can also be useful to practice text processing with publicly available online texts. In this section, we will collect text from **Wikipedia** using a method called **web scraping**. **Web scraping** refers to the automated process of retrieving content from web pages using a script instead of manually copying and pasting text. In Python, this is often done by sending a request to a website and downloading the HTML content of the page.

For this example, we will use the Python library **`requests`**, which allows us to send simple HTTP requests and retrieve the content of a web page. This is one of the easiest ways to access text data from the web. 

> Note: For larger or **more complex scraping projects,** researchers often use more advanced tools such as **Selenium**, which can interact with dynamic websites and simulate user actions like clicking or scrolling.

>Wikipedia is a convenient source for learning web scraping because the **Wikimedia Foundation** explicitly allows automated access to its content under certain conditions. The texts are openly licensed, and the site structure is relatively stable and well documented. However, this does not apply to all websites. Many websites restrict automated scraping in their **Terms of Service**, and in some cases **legal or ethical considerations must be taken into account.** Researchers should therefore always check whether scraping is permitted before collecting data from a website.

In the next step, we will send a request to a Wikipedia page and download its content so that we can extract the text for our small example corpus.

## 2.1 Scrape the Wikipedia website of the State of Bremen

In this first step, we will scrape the English-speaking website of the State of Bremen: https://en.wikipedia.org/wiki/Bremen_(state)

To do this, we use the **`requests`** library to download the HTML content of the webpage.  

````python
# Import the 'requests' and 'os' libraries
import requests       # To fetch the webpage
import os             # To create folders and work with file paths
````

````python
# Define the URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Bremen_(state)"
````

````python
# Wikipedia asks scripts to send a 'User-Agent' string
# This simply makes it transparent for the homepage provider who wants to scrape their page
headers = {"User-Agent": "PythonWorkshop/1.0 (educational example)"}

# Send a request to the page and get the HTML content
response = requests.get(url, headers=headers)

# Define the folder where the raw HTML files will be stored
output_folder = "../Data/DataA/web-scraping/raw"

# Create the folder if it does not exist yet
os.makedirs(output_folder, exist_ok=True)

# Define the output path for the raw HTML file
output_file = "../Data/DataA/web-scraping/raw/bremen_state.html"

# Save the raw HTML to a file
with open(output_file, "w", encoding="utf-8") as f:
    f.write(response.text)

# Print status code; helpful for troubleshooting
print("Status code:", response.status_code)

# Print a short preview of the HTML
print(response.text[:1000])
````

### Note on HTML Data

Most webpages on the internet are written in **HTML (HyperText Markup Language)**.  
HTML is a markup language that structures the content of a webpage, such as headings, paragraphs, links, images, and navigation menus.

You can often recognize HTML by the **tags enclosed in angle brackets**, for example:

```html
<p>This is a paragraph.</p>
<a href="https://example.com">This is a link</a>

These tags tell the browser how the content should be displayed, but they are not part of the actual text we want to analyze.

When we scrape a webpage, we usually download the entire HTML document, which includes both the visible text and all the structural markup.
Before we can perform text analysis, we therefore need to extract the readable text and remove the HTML tags.

In the next notebook, we will use a Python library called BeautifulSoup to parse the HTML and extract the relevant text from the webpages we downloaded.
